```python
# Region
Western = ['Luanda','Bondo','Butere','Siaya','Busia','Serem','Kakamega','Bumala']
Eastern = ['oloitoktok','Kibwezi','Matuu','Kitui','Kajiado','Machakos','Wote']
Nyanza = ['Mbita','Kisumu','Katito','Masara','Homa Bay','Kendu Bay','Rongo','Ndhiwa','Magunga']
Southrift = ['Narok','Chepsoen','Kipsitet','Nyangusu','Bomet','Oyugis','Nyamira']
Northrift = ['Bungoma','Eldoret','Kakuma','Kapenguria','Kapsabet','Kipkaren','Kitale','Kapsowar','Lodwar']
Coast = ['Kilifi','Kwale','Malindi','Kinango','Voi','Taveta','Hola']
Central = ['Kabarnet','Muranga','Nakuru','Nyahururu','Tharaka Nithi','Maralal']
```

```python
def region(shop):
    if shop in Western:
        return "Western"
    elif shop in Eastern:
        return "Eastern"
    elif shop in Nyanza:
        return "Nyanza"
    elif shop in Southrift:
        return "Southrift"
    elif shop in Northrift:
        return "Northrift"
    elif shop in Coast:
        return "Coast"
    elif shop in Central:
        return "Central"
    else:
        return "Verify"
    
sales_data["region"] = sales_data["shop_id/name"].apply(region)
sales_data.head()
```

```python
df = sales_data.groupby(['region','shop_id/name']).down_payment_paid_date.count()
df = pd.DataFrame(df)
df
```

```python
Zone_1 = ['Nyanza','Southrift','Northrift','Western']
Zone_2 = ['Coast','Central','Eastern']
def zone(region):
    if region in Zone_1:
        return "Zone_1"
    elif region in Zone_2:
        return "Zone_2"
    else:
        return "Verify"
    
sales_data['Zone'] = sales_data['region'].apply(zone)
# sales_data.query("Zone == 'Verify'")
sales_data
df2 = sales_data.groupby(['Zone','region','shop_id/name']).down_payment_paid_date.count()
df2 = pd.DataFrame(df2)
df2
```
